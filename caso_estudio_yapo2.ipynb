5.1.- ¿Cuál es el precio promedio de los anuncios por categoría?

```python
cursor.execute('SELECT categoria, AVG(precio) AS promedio FROM anuncio WHERE precio != \'N/A\' GROUP BY categoria;')
```

SELECT categoria, AVG(precio) AS promedio<br>
FROM anuncio<br>
WHERE precio != 'N/A'<br>
GROUP BY categoria;

5.2.- ¿Cuál es el número de anuncios por vendedor y por día?

```python
cursor.execute('SELECT id_vendedor, fecha_publicacion, count(*) AS anuncios FROM anuncio GROUP BY fecha_publicacion, id_vendedor;')
```

SELECT id_vendedor, fecha_publicacion, count(\*) AS anuncios<br>
FROM anuncio<br>
GROUP BY fecha_publicacion, id_vendedor;

5.4.-¿Cuál es el precio promedio de los anuncios según el día de la semana (lunes, martes, miercoles, etc.)?


```python
cursor.execute('SELECT WEEKDAY(fecha_publicacion) AS dia, AVG(precio) AS promedio FROM anuncio WHERE precio != \'N/A\' GROUP BY dia;')
```

SELECT WEEKDAY(fecha) AS dia, AVG(precio) AS promedio<br>
FROM anuncio<br>
WHERE precio != 'N/A'
GROUP BY WEEKDAY(fecha_publicacion);
