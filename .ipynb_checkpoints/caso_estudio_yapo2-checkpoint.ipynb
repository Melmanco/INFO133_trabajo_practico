5.1- ¿Cuál es el precio promedio de los anuncios por categoría?

```python
cursor.execute('SELECT categoria, AVG(precio) AS promedio FROM anuncio WHERE precio != \'N/A\' GROUP BY categoria;')
```

SELECT categoria, AVG(precio) AS promedio<br>
FROM anuncio<br>
WHERE precio != 'N/A'<br>
GROUP BY categoria;

5.2- ¿Cuál es el número de anuncios por vendedor y por día?

```python
cursor.execute('SELECT id_vendedor, fecha_publicacion, count(*) AS anuncios FROM anuncio GROUP BY fecha_publicacion, id_vendedor;')
```

SELECT id_vendedor, fecha_publicacion, count(\*) AS anuncios<br>
FROM anuncio<br>
GROUP BY fecha_publicacion, id_vendedor;

```python
df1 = DataFrame(cursor.fetchall())
df1.columns = cursor.column_names
df1
```

5.4-¿Cuál es el precio promedio de los anuncios según el día de la semana (lunes, martes, miercoles, etc.)?


```python
cursor.execute('SELECT WEEKDAY(fecha_publicacion) AS dia, AVG(precio) AS promedio FROM anuncio WHERE precio != \'N/A\' GROUP BY dia;')
```

SELECT WEEKDAY(fecha) AS dia, AVG(precio) AS promedio<br>
FROM anuncio<br>
WHERE precio != 'N/A'
GROUP BY WEEKDAY(fecha_publicacion);

```python
df1 = DataFrame(cursor.fetchall())
df1.columns = cursor.column_names
df1
```

8- Implementar un script Python para mostrar las 10 palabras claves (“más frecuentes”) de cada categoría de anuncios

Creamos un iterator con la misma cantidad de elementos que largo posible del título, para poder recorrerlo

```python
cursor.execute('CREATE TABLE iterator (char_index TINYINT UNSIGNED);')
for i in range(1,202):
    cursor.execute('INSERT INTO iterator (char_index) VALUES ({})'.format(i))
```

```python
cursor.execute('SELECT * FROM (SELECT SUBSTRING_INDEX(SUBSTRING_INDEX(a.titulo, \' \', i.char_index), \' \', -1) AS palabra, categoria, COUNT(*) AS ocurrencias, CASE WHEN @categoria != a.categoria THEN @filanum := 1 ELSE @filanum := @filanum + 1 END AS rango FROM anuncio AS a JOIN iterator AS i ON i.char_index <= length(a.titulo) - length(replace(a.titulo, \' \', \'\')) + 1 GROUP BY palabra, categoria ORDER BY categoria) q WHERE rango <= 10 ORDER BY categoria;')
```

SELECT<br>
&emsp;SUBSTRING_INDEX(SUBSTRING_INDEX(a.titulo, \' \', i.char_index), \' \', -1) AS palabra,<br>
&emsp;COUNT(\*) AS ocurrencias<br>
FROM anuncio AS a JOIN iterator AS i<br>
ON i.char_index <= length(a.titulo) - length(replace(a.titulo, \' \', \'\')) + 1<br>
GROUP BY palabra<br>
ORDER BY ocurrencias<br>
DESC LIMIT 10;
